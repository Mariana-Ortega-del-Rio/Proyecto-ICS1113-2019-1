## 2. Formulaci√≥n del modelo
### Conjuntos
- $T$: Conjunto de periodos de tiempo (en meses) donde $t_0 \in T$ es el primer mes y $t_5 \in T$ es el √∫ltimo mes.
- $P$: Conjunto de producto
- $M$: Conjunto de m√°quinas

### Par√°metros

- Para cada productp $p \in P$ y cada tipo de m√°quina $m \in M$ se nos da el tiempo $f_{p,m}$ (en horas) que el producto $p \in P$ necesita para ser manufacturado en la m√°quina $m \in M$
- Para cada mes $t \in T$ y cada producto $p \in P$ es conocido el l√≠mite superior de ventas $l_{t,p}$ para ese producto en ese mes
- Para cada producto se conoce su ganancia $k_p$
- Para cada mes $t \in T$ y cada m√°quina $m \in M$ es conocido el n√∫mero de m√°quinas disponibles $q_{t,m}$ 
- Cada m√°quina puede trabajar $g$ horas al mes
- Pueden haber m√°ximo $z$ productos de cada tipo guardados en cada mes y el costo de almacenarlo es $r$ por producto por mes.
- El √∫ltimo mes deben haber $w$ productos de cada tipo almacenados

### Variables
- $b_{t,p}$ describe la cantidad que se produce de cada producto $p$ en el mes $t \in T$
- $u_{t,p}$ describe la cantidad que se vende de cada producto $p \in P$ en el mes $t \in T$
- $s_{t,p}$ describe la cantidad que se almacena del producto $p \in P$ en el mes $t \in T$

  $b_{t,p}, u_{t,p}, s_{t,p} \geq 0 \forall t \in T, \forall p \in P$

### Funci√≥n objetivo
$$ max \sum_{t \in T} \sum_{p \in P} (k_p \cdot u_{t,p} - r \cdot s_{t,p}) $$


- Restricci√≥n de inventario

$$s_{t-1,p} + b_{t,p} = u_{t,p} + s _{t,p} \; \forall \; t \in T \setminus t_o \; \forall \; p \in P$$


&nbsp;
$$b_{t,p} = u_{t,p} + s_{t,p} \; \forall \; p \in P$$


&nbsp;
- Al final del √∫ltimo mes debe haber cierta cantidad almacenada de cada producto


&nbsp;
$$ s_{t_5,p} = w \; \forall \; p \in P, \; \forall \; t_{5} \in T$$

- Capacidad de almacenaje


&nbsp;
$$s_{t,p} \leq z \; \forall \; p \in P, \; \forall \; t \in T$$


&nbsp;
- El tiempo total en cada m√°quina por mes debe ser menor o igual que la cantidad de horas disponibles para esa m√°quina en ese mes.
